```python
from dbfread import DBF
from pandas import DataFrame
```

```python
path_parts = 'C:\\Base\\tovar.dbf'
path_inv = 'C:\\Base\\skladt.dbf'
dict_parts = DBF(path_parts,encoding="1251",ignore_missing_memofile=True,ignorecase=False)
dict_inv = DBF(path_inv, encoding="1251", ignore_missing_memofile=True,ignorecase=False)
```

```python
parts_frame = DataFrame(iter(dict_parts))
inv_frame = DataFrame(iter(dict_inv))
```

```python
parts_col = ['ART', 'C_TOV', 'TOV_SNAME', 'TOV_CENR1', 'TOV_UPAK', 'TOV_ED', 'PROIZV','IZGOT']
parts_frame = parts_frame[parts_col]
idx = parts_frame['C_TOV']
parts_frame = DataFrame(parts_frame, index=idx)
parts_frame.head(6)
```

```python
inv_col = ['C_TOV', 'TOV_KOL', 'TOV_REZ']
inv_frame = inv_frame[inv_col]
inv_frame = inv_frame.groupby(['C_TOV']).sum()
inv_frame.head()
```

```python
m = parts_frame.merge(inv_frame, how='left')
m.fillna(0, inplace=True)
m.head(6)
```
